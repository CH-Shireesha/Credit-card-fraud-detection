```python
neighbours = np.arange(1,25)
train_accuracy =np.empty(len(neighbours))
test_accuracy = np.empty(len(neighbours))

for i,k in enumerate(neighbours):
    #Setup a knn classifier with k neighbors
    knn=KNN(n_neighbors=k,algorithm="kd_tree",n_jobs=-1)
    
    #Fit the model
    knn.fit(train_x,train_y.ravel())
    
    #Compute accuracy on the training set
    train_accuracy[i] = knn.score(train_x,train_y.ravel())
    
    #Compute accuracy on the test set
    test_accuracy[i] = knn.score(test_x, test_y.ravel()) 
```

```python
idx = np.where(test_accuracy == max(test_accuracy))
x = neighbours[idx]
```

```python
# Creating instance of KNN
knn = KNN(n_neighbors=x[0],algorithm="kd_tree",n_jobs=-1)

# Fitting the model
knn.fit(train_x, train_y.ravel())

# Predicting over the Train Set and calculating F1
train_predi = knn.predict(train_x)
k1 = f1_score(train_y,train_predi)
print('Train F1 Score    ', k1 )
```

```python
# Predicting over the Test Set and calculating F1
test_predi = knn.predict(test_x)
k1 = f1_score(test_y,test_predi)
print('Test F1 Score    ', k1 )
```
